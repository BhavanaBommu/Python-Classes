```python
# Generator function for the cube of numbers (power of 3)
def gencubes(n):
    for num in range(n):
        yield num**3
```

```python
gencubes(10)
for x in gencubes(9):
    print(x)
```

```python
def genfibon(n):
    '''
    Generate a fibonacci sequence up to n
    '''
    a = 1
    b = 1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
genfibon(5)
for num in genfibon(12):
    print(num)
```

```python
#object with __iter__() and __next__() methods is an iterator.
def simple_gen():
    for x in range(3,9):
        yield x
```

```python
# Assign simple_gen 
g = simple_gen()
g
print(next(g))
```

```python
print(next(g))
```

```python

def fahrenheit(T):
    return ((float(9)/5)*T + 32)


def celsius(T):
    return (float(5)/9)*(T-32)
    
temp = [0, 22.5, 40,100]


def summ(x, y ):
    return x+y


for i in temp:
    celsius =  (float(5)/9)*(i-32)
    print(celsius)
```

```python
F_temps = list(map(celsius, temp))

#Show
F_temps
```

```python
# Convert back
list(map(fahrenheit, F_temps))
```
